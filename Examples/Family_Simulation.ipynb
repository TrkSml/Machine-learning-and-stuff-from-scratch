# Writing our main function

```python
# __ author __ = Tarek Samaali

import numpy as np

def simulate(N,number_of_boys):
    List_of_couples=[[] for _ in range(N)]
    FLAG=[False for i in range(N)]

    while not all(FLAG):
        for i in range(len(List_of_couples)):
            item=np.random.random()
            if not FLAG[i]:
                #test if it is a girl
                #Equiprobability to have either a boy or a girl
                if item>.5 :
                    List_of_couples[i]+=[0]
                #Then it is a boy
                else :
                    List_of_couples[i]+=[1]
                    if List_of_couples[i].count(1)==number_of_boys:
                        FLAG[i]=True

    #Long-term Ratio
    Ratio_parameter=reduce(lambda x,y:x+y,List_of_couples)
    Final_Ratio=sum(Ratio_parameter)/float(len(Ratio_parameter)-sum(Ratio_parameter))
    
    return Final_Ratio

```

# Showing the result !

```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

list_number_of_boys=[1,5,10]
lower_bound_population=10
upper_bound_population=1000

X1=range(lower_bound_population,upper_bound_population,1)
Y=[map(lambda x:simulate(x,number_of_boys),X1) for number_of_boys in list_number_of_boys]
Y_colors=['r','b','g']

plt.figure(figsize=(10,10))
params = {'legend.fontsize': 20,
          'legend.handlelength': 20}

plt.rcParams.update(params)

for simulation_index in range(len(Y)):
    plt.plot(X1,Y[simulation_index],Y_colors[simulation_index])

plt.xlabel("Number of families")

plt.ylabel("Ratio Boys/Girls")

plt.show()

## The long-term ratio of boys to girls in the country is equal to 1
```
